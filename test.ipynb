```python
import pandas as pd
import numpy as np
```

```python
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 22, 35, 28]
}

# 使用字典创建 DataFrame
df = pd.DataFrame(data)
```

```python
# Utils工具类
def convertCode(code) -> str:
    """
    将股票代码转换为标准6位
    :param code: 股票代码
    :return: 6位标准代码
    """
    return "{:06d}".format(int(code)) #用于将整数 code 转换为一个包含6位数字的字符串 不足6位的部分在左侧用0填充
```

```python
df[np.isin(df['ID'].apply(convertCode).str[:3], ['000', '600', '601', '603', '605'])]
```

```python
df.groupby(['ID']).apply(lambda x: pd.DataFrame(
        {
            'Name': ['Alice', 'Bob'],
            'Age': [
                x[x['Name'] == 'Alice']['Age'].iat[0] if len(x[x['Name'] == 'Alice']['Age']) > 0 else 0,
                x[x['Name'] == 'Bob']['Age'].iat[0] if len(x[x['Name'] == 'Bob']['Age']) > 0 else 0
            ]
        }).dropna()).reset_index()#.drop(['level_1'], axis=1)
```

```python
import numpy as np

# 定义普通的 Python 函数
def add_five(x):
    return x + 5

# 使用 np.vectorize 转换为向量化函数
vectorized_add_five = np.vectorize(add_five)

# 创建一个 NumPy 数组
arr = np.array([1, 2, 3, 4, 5])

# 使用向量化函数操作数组
result = vectorized_add_five(arr)

print(result)
```
